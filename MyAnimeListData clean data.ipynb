```python
# determining if there exists unaired values
anime_info_raw['aired'].value_counts() # contains null values
# {'from': None, 'to': None}  
anime_info_raw.loc[(anime_info_raw['aired']== "{'from': None, 'to': None}") & (anime_info_raw['status'] == "Not yet aired"),['anime_id','title','status','aired']].info() 
anime_info_raw.loc[(anime_info_raw['status'] == "Not yet aired"),['anime_id','title','status','aired']] # anime airing date has been announced

# using the status column is a better way to filter unaired anime = those that are still in the works, not yet premiered with only info like production videos, teasers, and the OP having been released  
```

#### Analyis: 
No. of rows with status = 'not aired yet' and no aired info ({'from': None, 'to': None}) = 147
whereas the no. of rows with status = 'not aired yet' is much higher at 336 rows -> on inspection, these manga have been completed and their airing info has been announced by the distributor. since we do not have info like score for these manga -> using the condition status = 'not aired yet' is a better filter
